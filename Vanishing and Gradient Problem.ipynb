# Vanishing Gradient & Exploading Gradient

1. Sigmoid Activation Funciton.(derivative of sigmoid is between 0 to 0.25)
2. Tanget(Tanh) Activation Fucntion.(derivative of Tanh is between 0 to 1)

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Generate input values
x = np.linspace(-10, 10, 100)

# Calculate sigmoid values
sigmoid_values = sigmoid(x)

# Calculate derivative (gradient) values
gradient_values = sigmoid_derivative(x)

# Plot sigmoid function and its derivative
plt.figure(figsize=(10, 5))

# Plot sigmoid function
plt.plot(x, sigmoid_values, label='Sigmoid', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)

# Plot derivative of sigmoid function
plt.plot(x, gradient_values, label='Derivative', linestyle='--', color='red')
plt.scatter([0], [0.25], color='green', label='Vanishing Gradient Point')  # Mark point where gradient becomes very small
plt.title('Derivative of Sigmoid Function (Gradient)')
plt.xlabel('Input')
plt.ylabel('Gradient')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Observation

    From the image we can see the the derivate id 0.25 . 

    Vanishing gradient happens when this value become so small if the model keep on iterrating because of chain rule and at some point the old weight and new weight will be same allowing the model not to update weight further
    
***Tanh Fucntion***

For Tanh the derivate is between 0 and 1


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of weight values
weights = np.linspace(-5, 5, 100)

# Define a simple loss function (for illustration purposes)
loss = weights**2  # Quadratic loss function

# Calculate the gradient of the loss function with respect to the weight parameter
gradient = 2 * weights  # Gradient of the quadratic loss function

# Plotting
plt.plot(weights, loss, label='Loss Function (e.g., Quadratic)')
plt.plot(weights, gradient, linestyle='--', color='green', label='Gradient')
plt.scatter(0, 0, color='red', label='Initial Weight')  # Initial weight at (0, 0)

# Mark the point where the gradients become excessively large (exploding gradient)
diverge_index = np.argmax(np.abs(gradient) > 10)  # Example threshold for divergence
plt.scatter(weights[diverge_index], gradient[diverge_index], color='orange', label='Divergence Point')

# Add labels and legend
plt.xlabel('Weight')
plt.ylabel('Loss / Gradient')
plt.title('Exploding Gradient Problem (Simplified)')
plt.legend()
plt.grid(True)
plt.show()
```

## Observatoin

    Happens because of large weight initilizatoin and while backpropagation this weights will become morelarge and difficult to converge.
    
     Both in sigmoid and Tanh functions are not robust to Exploading gradient because the output of this functions derivaties are sometimes large potive or negative values for large weights
