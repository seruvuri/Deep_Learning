## Non-Zero Centererd 

The output of activation functions are not centerd to zero i.e mean value of output is not close to zero .

IN Zero centrerd the ouput value has either positive or negative values which helps optimizers to update weight in one directions

### Advantages

Due to weight update in one direction the loss function convergence will be slower.

## Activation functoins
1. Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of input values
x = np.linspace(-5, 5, 100)

# Define the activation functions
sigmoid = 1 / (1 + np.exp(-x))
#tanh = np.tanh(x)

# Plot the activation functions
plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid, label='Sigmoid', color='blue')
#plt.plot(x, tanh, label='Tanh', color='red')
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)  # Horizontal line at y=0
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)  # Vertical line at x=0
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions with Non-Zero-Centered Outputs')
plt.legend()
plt.grid(True)
plt.show()
```

## Weight Update in Zero Centered Activatoin Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the weight range
weights = np.linspace(-5, 5, 100)

# Define a simplified quadratic loss function
loss = weights**2

# Plot the loss function
plt.plot(weights, loss, label='Loss Function')

# Mark the initial weight position
initial_weight = 3
plt.scatter(initial_weight, initial_weight**2, color='red', label='Initial Weight')

# Example positive and negative weight updates
positive_update = 1
negative_update = -1

# Plot the positive weight update
plt.arrow(initial_weight, initial_weight**2, positive_update, (initial_weight + positive_update)**2 - initial_weight**2,
          head_width=0.2, color='green', label='Positive Weight Update')

# Plot the negative weight update
plt.arrow(initial_weight, initial_weight**2, negative_update, (initial_weight + negative_update)**2 - initial_weight**2,
          head_width=0.2, color='blue', label='Negative Weight Update')

plt.xlabel('Weight')
plt.ylabel('Loss')
plt.title('Zero centerd weight update')
plt.legend()
plt.grid(True)
plt.show()
```

## Zero Centererd 

The output of activation functions are centerd to zero i.e mean value of output is close to zero .

IN Zero centrerd the ouput value has both positive and negative values which helps optimizers to update weight in both directions

### Advantages

Due to weight update in both(positive and negative) directions loss function convergence will be fast
